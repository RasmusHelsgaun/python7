# 1. Find the top 10 players with the most kills on origin/pc

```python
import bs4

with open('./kills.html') as f:
    kills_html = f.read()

soup = bs4.BeautifulSoup(kills_html)

bodyList = soup.select("tbody tr")
top_10 = {}

for i in range(10):
    rank = bodyList[i].select("td")[0].getText()
    name = bodyList[i].select("td")[1].select("a")[0].getText()
    kills = bodyList[i].select("td")[2].getText()
    kills = kills.replace(',', '')
    top_10[f"{rank} - {name}"] = int(kills)

top_10
```

```python
%pylab inline
plt = matplotlib.pyplot

plt.figure(figsize=(15,8))
plt.title("Top 10 players with most Kills in Apex")
plt.xlabel("Rank / Name")
plt.ylabel("Kills")
           
names = top_10.keys()
kills = top_10.values()

plt.bar(names, kills)

plt.xticks(rotation=45)

plt.show()
```

# 2. For each weapon type find the player with the most kills with that weapon

```python
def get_top_1_weapon_kills(html_files):
    res = {}
    for i in range(len(html_files)):
        with open(f'./{html_files[i]}.html') as f:
            weapon_html = f.read()

        soup = bs4.BeautifulSoup(weapon_html)
        
        weapon_title = soup.select("div div section h1 span")[0].getText()
        weapon = weapon_title.split()[0]
    
        bodyList = soup.select("tbody tr td")
        
        name = bodyList[1].select("a")[0].getText()
        
        kills = bodyList[2].getText()
        kills = kills.replace(',', '')
        
        res[f"{weapon} - {name}"] = int(kills)
        
    return res
```

```python
weapon_files = ["pistol", "shotgun", "smg", "ar", "lmg", "sniper"]
top_1s = get_top_1_weapon_kills(weapon_files)
top_1s
```

```python
plt.figure(figsize=(15,8))
plt.title("Top weapon kills by players in Apex")
plt.xlabel("Weapon and top player")
plt.ylabel("Kills")
           
wep_player = top_1s.keys()
kills = top_1s.values()

plt.bar(wep_player, kills)

plt.xticks(rotation=45)

plt.show()
```
